```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('amazon_sales.csv')  # Replace with your actual filename

# Clean price columns: remove ₹ and commas
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)

# Convert to numeric
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Drop NaNs
df = df.dropna(subset=['actual_price', 'discounted_price'])
df = df.drop_duplicates()

print("Cleaned shape:", df.shape)
print(df[['actual_price', 'discounted_price']].head())

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='actual_price',
    y='discounted_price',
    data=df,
    alpha=0.6,
    edgecolor='white',
    linewidth=0.5
)
plt.title("Actual Price vs Discounted Price", fontsize=14, pad=15)
plt.xlabel("Actual Price (₹)", fontsize=12, labelpad=10)
plt.ylabel("Discounted Price (₹)", fontsize=12, labelpad=10)
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout(pad=2)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('amazon_sales.csv')  # Replace with your actual file name

# Clean and convert necessary columns
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False)

# Convert all to numeric (handle '|', missing, etc.)
cols_to_convert = ['actual_price', 'discounted_price', 'rating', 'rating_count', 'discount_percentage']
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop rows with any NaNs in selected columns
df_corr = df[cols_to_convert].dropna()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df_corr.corr(), annot=True, cmap="coolwarm", linewidths=0.5, linecolor='white')
plt.title("Feature Correlation Heatmap", fontsize=14)
plt.tight_layout()
plt.show()
```

#### 7. Pie chart: discount % vs rating

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load dataset
df = pd.read_csv("amazon_sales.csv")

# Step 2: Clean discount_percentage column
# Remove % sign if it exists and convert to float
df['discount_percentage'] = (
    df['discount_percentage']
    .astype(str)  # convert to string for cleaning
    .str.replace('%', '', regex=False)
)

# Step 3: Convert cleaned values to float
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Step 4: Drop rows with NaN in discount_percentage
df = df.dropna(subset=['discount_percentage'])

# Step 5: Create discount bins (FLOAT safe)
df['discount_bin'] = pd.cut(
    df['discount_percentage'],
    bins=[0, 20, 40, 60, 80, 100],
    labels=["0-20%", "21-40%", "41-60%", "61-80%", "81-100%"]
)

# Step 6: Count values in each bin
discount_dist = df['discount_bin'].value_counts().sort_index()

# Step 7: Plot pie chart
plt.figure(figsize=(7, 7))
colors = sns.color_palette('Set2')
discount_dist.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'black'}
)

plt.ylabel('')
plt.title("Product Distribution by Discount Range")
plt.tight_layout()
plt.show()
```
